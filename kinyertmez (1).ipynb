```python
import pandas as pd
```

```python
df1 = pd.read_excel('kinyert méz 2000-2005.xls')
```

```python
df1.rename(columns={'14.14 PRODUCŢIA AGRICOLĂ ANIMALĂ': 'Producţia agricolă animală'}, inplace=True)
```

```python
df2 = pd.read_excel('kinyert méz 2006-2011.xls')
```

```python
df2.rename(columns={'14.17   Producţia agricolă animală': 'Producţia agricolă animală'  }, inplace=True)
```

```python
df3 = pd.read_excel('kinyert méz 2012-2017.xls')
```

```python
df3.rename(columns={'14.18   Producţia agricolă animală': 'Producţia agricolă animală' }, inplace=True)
```

```python
df4 = pd.read_excel('kinyert méz 2018-2020.xls')
```

```python
df4.rename(columns={'14.16   Producţia agricolă animală': 'Producţia agricolă animală' }, inplace=True)
```

```python
df5 = pd.read_excel('kinyert méz 2021-2022.xls')
```

```python
df5.rename(columns={'14.16   Producţia agricolă animală': 'Producţia agricolă animală' }, inplace=True)
```

```python
mdf=pd.merge(df1,df2, on=('Animale ce revin la 100 hectare teren'), how='inner')
```

```python
mdf.rename(columns={'Animale ce revin la 100 hectare teren': 'Producţia agricolă animală'}, inplace=True)
```

```python
mdf=pd.merge(mdf,df4, on=('Producţia agricolă animală'), how='inner')
```

```python
mdf=pd.merge(mdf,df5, on=('Producţia agricolă animală'), how='inner')
```

az elso 3 tablazatot osszeteszi majd kitorli a teljesen ures sorokat

```python
cdf1 = pd.concat([df1, df2, df3], axis=1)
cdf1 = cdf1.dropna(axis=1, how='all')
cdf1=cdf1.dropna(axis=0, how='all')
```

kivalogatja, hogy csak a lenyeges sorok maradjanak

```python
mez_00_17 = pd.concat([cdf1.iloc[1:3], cdf1.iloc[30:31]])
```

a din care: felirat utan odahelyezi a cella jo tartalmat

```python
for i, row in mez_00_17.iterrows():
    for j, cell in enumerate(row):
        if cell == 'din care:':
            row[j] += " proprietate majoritar privata"
```

```python
cdf2 = pd.concat([df4, df5], axis=1)
cdf2 = cdf2.dropna(axis=1, how='all')
cdf2 = cdf2.dropna(axis=0, how='all')
```

```python
mez_18_22 = pd.concat([cdf2.iloc[1:3], cdf2.iloc[50:51]])
```

```python
for i, row in mez_18_22.iterrows():
    for j, cell in enumerate(row):
        if cell == 'din care:':
            row[j] += " proprietate majoritar privata"
```

```python
mez_00_22 = pd.concat([mez_00_17.reset_index(drop=True), mez_18_22.reset_index(drop=True)], axis=1)
```

```python
mez_00_22
```

megfordítja a táblázatot

```python
mez_megforditott = mez_00_22.transpose()
```

kitörli azokat a sorokat ahol az utolso oszlopban nem szám szerepel

```python
mez_megforditott = mez_megforditott[pd.to_numeric(mez_megforditott[2], errors='coerce').notna()]
```

átírja a szoveget a kivalasztott cellában (0-tol számolja)

```python
mez_megforditott.iloc[8, 0] = "2004"#2005re is
```

```python
mez_megforditott['Év'] = pd.to_numeric(mez_megforditott['Év'])
```

```python
mez_megforditott[0].fillna(method='ffill', inplace=True)
```

```python
mez_megforditott.loc[mez_megforditott[1] == 'din care:', 1] = 'ebből: többnyire magántulajdon'
```

```python
mez_megforditott.rename(columns={0: 'Év'}, inplace=True)
mez_megforditott.rename(columns={1: 'Össz'}, inplace=True)
mez_megforditott.rename(columns={2: 'Kinyert méz (tonna)'}, inplace=True)
mez_megforditott.reset_index(drop=True, inplace=True)
```

```python
mez_megforditott.to_excel('kinyertmez.xlsx')
```
