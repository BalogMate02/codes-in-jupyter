```python
import pandas as pd
```

```python
df1 = pd.read_excel('méhcsaládok 2000-2005.xls')
df2 = pd.read_excel('méhcsaládok 2001-2006.xls')
df3 = pd.read_excel('méhcsaládok 2007-2012.xls')
df4 = pd.read_excel('méhcsaládok 2013-2018.xls')
df5 = pd.read_excel('méhcsaládok 2019-2020.xls')
df6 = pd.read_excel('méhcsaládok 2021-2022.xls')
```

```python
cdf1 = pd.concat([df1, df2, df3, df4, df5, df6], axis=1)
cdf1 = cdf1.dropna(axis=1, how='all')
cdf1=cdf1.dropna(axis=0, how='all')
```

```python
mehcsalad = pd.concat([cdf1.iloc[4:6], cdf1.iloc[29:30]])
```

```python
mehcsalad = mehcsalad.transpose()
```

```python
mehcsalad.rename(columns={4: 'Év'}, inplace=True)
mehcsalad.rename(columns={5: 'Össz'}, inplace=True)
mehcsalad.rename(columns={30: 'Méhcsalád (ezer)'}, inplace=True)
mehcsalad.reset_index(drop=True, inplace=True)
```

```python
mehcsalad = mehcsalad[pd.to_numeric(mehcsalad['Méhcsalád (ezer)'], errors='coerce').notna()]
```

duplikalt sorok kitorlese

```python
duplicates = mehcsalad.duplicated(subset=['Év', 'Össz', 'Méhcsalád (ezer)'])
mehcsalad = mehcsalad[~duplicates]
```

```python
mehcsalad.iloc[8, 0] = 2004
mehcsalad.iloc[10, 0] = 2005
mehcsalad.iloc[12, 0] = 2006
```

```python
mehcsalad['Év'].fillna(method='ffill', inplace=True)
mehcsalad.loc[mehcsalad['Össz'] == 'din care:', 'Össz'] = 'ebből: többnyire magántulajdon'
```

```python
mehcsalad.to_excel('mehcsalad.xlsx')
mehcsalad.to_csv('mehcsalad.csv')
```
